```python
#Loading data in dataframe, specify which fields to use 
fields = ["PERIDNUM","A_AGE","A_HGA","A_SEX","PECERT1","A_MJIND","PRUNTYPE","RSNNOTW","PEARNVAL","PTOT_R","UC_YN","A_EXPLF","A_LFSR"]
df = pd.read_csv("C:/Users/13013/OneDrive/CTEC426/pppub24.csv", sep = ",", usecols = fields)
df.head(30)
```

```python
print(df.shape)
```

```python
#Recoding Reason for Unemployment 
#PRUNTYPE: Reason for Unemployment
#0.NIU 1.Unemployed
df['PRUNTYPE'] = df['PRUNTYPE'].replace([1,2,3,4,5,6],1)

#A_HGA: Educational Attainment
#0.Children, 1.< HS Diploma, 2.HS Diploma, 3.Some college no degree, 4.Associate's, 5.Bachelors, 6.Masters, 7.Professional School, 8.Doctorate
df['A_HGA'] = df['A_HGA'].replace([31,32,33,34,35,36,37,38],1)
df['A_HGA'] = df['A_HGA'].replace([39],2)
df['A_HGA'] = df['A_HGA'].replace([40],3)
df['A_HGA'] = df['A_HGA'].replace([41,42],4)
df['A_HGA'] = df['A_HGA'].replace([43],5)
df['A_HGA'] = df['A_HGA'].replace([44],6)
df['A_HGA'] = df['A_HGA'].replace([45],7)
df['A_HGA'] = df['A_HGA'].replace([46],8)

#A_LFSR: Labor Force Status Recode: 0 = Children or Armed Forces, 1 = Working and/or With job not at work, 2 = Unemployed, looking for work and/or = Unemployed, on layoff, 3 = Nilf
df['A_LFSR'] = df['A_LFSR'].replace([1,2],1)
df['A_LFSR'] = df['A_LFSR'].replace([3,4],2)
df['A_LFSR'] = df['A_LFSR'].replace([7],3)

df.head(50)
```

```python
#Creating pie chart from A_LFSR: Labor Force Status Recode 
#Dropped values 0 and 3
sizes = df['A_LFSR'].value_counts().drop([0, 3])
labels = ['Employed','Unemployed']
#sizes = df['A_LFSR'].value_counts().sort_index()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
#Unemployment rate for the United States is 4%.
```
